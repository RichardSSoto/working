# Matplotlib

```python
#seaborn
#viz
import matplotlib.pyplot as plt
import numpy as np
```

# Ploteo Lineal

```python
#estética
plt.style.use('classic')
```

```python
x=np.linspace(0,10,100)
```

```python
plt.plot(x, np.sin(x))
plt.show()
```

# Dos gráficos en uno

```python
plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.figure()
#1er
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

#2do
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
#ploteo orienta a objetos
fig, ax = plt.subplots(2,1)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x, np.sin(x))
ax[0][1].plot(x, np.cos(x))
ax[1][0].plot(x, np.tan(x))
ax[1][1].plot(x, np.log(x))
```

# Customazing lines

```python
#estética
plt.style.use('seaborn-whitegrid')
```

```python
plt.figure()
plt.plot(x, np.sin(x-0), 'g-.')
plt.plot(x, np.sin(x-1), 'b--')
plt.plot(x, np.sin(x-2), 'r*')
plt.plot(x, np.sin(x-3), 'k:')
plt.show()
```

# Ploteo con leyendas

```python
y=np.sin(x[:,np.newaxis]+np.pi*np.arange(0,2,0.5))
```

```python
plt.plot(x,y[:,0],'r:',label='firts')
plt.plot(x,y[:,1],'g--', label='second')
plt.legend()
plt.xlabel('time')
plt.ylabel('sen(x)')
plt.title('Linea senos')
plt.show()
```

# Zoom

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('time')
axes1.set_ylabel('y')
axes1.set_title('Axes 1 title')

axes2.plot(y,x,'r')
axes2.set_xlabel('time')
axes2.set_ylabel('y')
axes2.set_title('Axes 2 title')

plt.show()
```

# Scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,30)
y=np.sin(x)
```

```python
plt.scatter(x,y,marker='+')
plt.ylim(-1.2,1.2)
plt.xlabel('time')
plt.ylabel('sen(x)')
plt.title('Linea senos')
plt.show()
```

```python
#nueva data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color=rng.randn(100)
sizes=1000*rng.randn(100)
```

```python
plt.scatter(x,y, c=color, s=sizes, alpha=0.3, cmap='plasma')
plt.colorbar()
plt.xlabel('Distancia')
plt.ylabel('Región')
plt.title('Scatter densidad')
plt.show()
```

# Error bars

```python
#nueva data
x=np.linspace(0,10,50)
dy=0.8
y=np.sin(x)+dy*np.random.randn(50)
```

```python
#plt.plot(x,y)
plt.errorbar(x,y,yerr=dy, fmt='.k')
```

```python
plt.errorbar(x,y,yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3)
plt.xlabel('Puntos')
plt.ylabel('Tics')
plt.title('Error plot')
plt.show()
```

# Contour plots

```python
def f(x,y):
  return np.sin(x)**10 + np.cos(10+x*y)*np.cos(x)
```

```python
x=np.linspace(0,5,50)
y=np.linspace(0,5,40)

X,Y=np.meshgrid(x,y)
Z=f(X,Y)
```

```python
plt.contour(X,Y,Z, colors='black')
```

```python
plt.contour(X,Y,Z, 20,cmap='RdGy')
```

```python
plt.contourf(X,Y,Z, 20,cmap='RdGy')
plt.colorbar()
```

# Histogramas

```python
data = np.random.randn(1000)
```

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=50, density=True, alpha=0.5, color='steelblue',edgecolor='none')
plt.xlabel('fecha')
plt.ylabel('duración')
plt.title('Histograma')
plt.show()
```

```python
#new data
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,2,1000)
```

```python
kwargs=dict(bins=50, density=True, alpha=0.3,edgecolor='none')
plt.figure(figsize=(10,5))
plt.hist(x1,**kwargs ,label='h1')
plt.hist(x2,**kwargs ,label='h2')
plt.hist(x3,**kwargs ,label='h3')

plt.legend()

plt.xlabel('fecha')
plt.ylabel('duración')
plt.title('Histograma')
plt.show()
```

# Boxplot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
plt.figure()
plt.boxplot(data, vert=False)

plt.xlabel('valores')
plt.ylabel('duración')
plt.title('Boxplot')
plt.show()
```

```python
plt.figure()
plt.boxplot(data, vert=True)

plt.xlabel('valores')
plt.ylabel('duración')
plt.title('Boxplot')
plt.show()
```
