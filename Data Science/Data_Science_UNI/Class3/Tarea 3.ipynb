# Matplotlib Exercises 


## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

Importa matplotlib.pyplot

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

## Ejercicio 1

* ** Crea un objeto figura llamado fig ---> plt.figure () **
* ** Use add_axes para agregar un eje al lienzo de la figura en [0,0,1,1]. Llame a este nuevo eje ax. ** **
* ** Plot (x, y) en esos ejes y configure las etiquetas y los títulos: **

```python
fig = plt.figure()
fig.add_axes([0,0,1,1], label='ax')
plt.plot(x,y,'b')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("title")
plt.show()
```

## Ejercicio 2

** Crea un objeto figura y coloca dos ejes sobre él, ax1 y ax2. Ubicado en [0,0,1,1] y [0.2,0.5, .2, .2] respectivamente. **

```python
fig = plt.figure()
fig.add_axes([0,0,1,1],label='ax1')
fig.add_axes([0.2,0.5, .2, .2],label='ax2')
fig.show()
```

** Ahora dibuje (x, y) en ambos ejes. Y llame a su figura objeto para mostrarlo. **

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1],label='ax1')
ax1.plot(x,y,'b')
ax1.set_xlabel('x1')
ax1.set_ylabel('y')
ax2=fig.add_axes([0.2,0.5, .2, .2],label='ax2')
ax2.plot(x,y,'b')
ax2.set_xlabel('x2')
ax2.set_ylabel('y')
fig.show()
```

## Ejercicio 3

** Cree el siguiente diagrama agregando dos ejes a un objeto de figura en [0,0,1,1] y [0.2,0.5, .4, .4] **

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5, .4, .4])
plt.show()
```

** Ahora use las matrices x, y, y z para recrear el diagrama a continuación. Observe los límites xlimits e y en el diagrama insertado: **

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,z)
ax2 = fig.add_axes([0.2,0.5, .4, .4])
ax2.plot(x,y)
plt.show()
```

## Ejercicio 4

** Use plt.subplots (nrows = 1, ncols = 2) para crear la gráfica a continuación. **

```python
fig, ax = plt.subplots(1,2)
```

** Ahora trace (x, y) y (x, z) en los ejes. Juega con el ancho de línea y el estilo **

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(x,y,'b--')
ax[1].plot(x,z,'r')
```

** Vea si puede cambiar el tamaño del diagrama agregando el argumento figsize () en plt.subplots () **

```python
fig, ax = plt.subplots(1,2,figsize=(12.5,2.5))
ax[0].plot(x,y,'b', linewidth=5)
ax[1].plot(x,z,'r--', linewidth=2)

```

```python
import numpy as np
import matplotlib.pyplot as plt

#CArgamos los datos
txt = np.loadtxt("Densidad_vs_masa_lambda_0.dat")
x1=txt[:,0]
y1=txt[:,1]
txt = np.loadtxt("Densidad_vs_masa_lambda_1.dat")
x2=txt[:,0]
y2=txt[:,1]
txt = np.loadtxt("Densidad_vs_masa_lambda_2.dat")
x3=txt[:,0]
y3=txt[:,1]
txt = np.loadtxt("Densidad_vs_masa_lambda_-1.dat")
x4=txt[:,0]
y4=txt[:,1]
txt = np.loadtxt("Densidad_vs_masa_lambda_-2.dat")
x5=txt[:,0]
y5=txt[:,1]


plt.style.use('seaborn-whitegrid')
plt.figure()
plt.plot(x1, y1, 'k', label= '$λ_{0}$')
plt.plot(x2, y2, 'b--',label= '$λ_{1}$')
plt.plot(x3, y3, 'r-.',label= '$λ_{2}$')
plt.plot(x4, y4, 'g-.',label= '$λ_{-1}$')
plt.plot(x5, y5, 'y:',label= '$λ_{-2}$')
plt.legend()
plt.title('Densidad vs Masa')
plt.xlabel('ρ[g/c$m^{3}$]')
plt.ylabel('M[$M_{0}$]')
plt.grid()
plt.show()
```
